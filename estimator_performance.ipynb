```python
import matplotlib.pyplot as plt
```

```python
est = []
act= []
f = open('estimations.csv')
for line in f:
    est += [eval(line)[0]]
    act += [eval(line)[1]]
```

```python
fig = plt.figure(figsize=(10,10))
plt.scatter([x for (x,y) in act], [y for (x,y) in act], color='blue', label='Gps-read')
plt.scatter([x for (x,y) in est], [y for (x,y) in est], color='red', alpha=.8, s=10, label='Estimator outputted')
plt.legend()
plt.plot()
```
